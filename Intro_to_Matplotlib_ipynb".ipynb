# Завдання
## 1. Створити Subplot із  4 графіків сінусоїди з різними частотами

1.1. Сформувати діапазон функції від 0 до 2 з кроком 0.01

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0,2,0.01)
```

1.2 Сформувати частоти сінусоїди

```python
s1 = np.sin(3 * np.pi * t)
s2 = np.sin(5 * np.pi * t)
s3 = np.sin(7 * np.pi * t)
s4 = np.sin(9 * np.pi * t)
```

1.3 Надрукувати графіки

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(t, s1, c='r',linewidth=4.0)
ax1.set(xlabel='Час (с)', ylabel='Частота', title='Сінус функція')
ax1.grid()

ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(t, s2, c='r',linewidth=4.0)
ax2.set(xlabel='Час (с)', ylabel='Частота', title='Сінус функція')
ax2.grid()

ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(t, s3, c='r',linewidth=4.0)
ax3.set(xlabel='Час (с)', ylabel='Частота', title='Сінус функція')
ax3.grid()

ax4 = fig.add_subplot(2, 2, 4)
ax4.plot(t, s4, c='r',linewidth=4.0)
ax4.set(xlabel='Час (с)', ylabel='Частота', title='Сінус функція')
ax4.grid()

plt.show()
```
