```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

1. Зчитати та завантажити тестове чорно-біле та кольорове зображення.

```python
img_bw = cv2.imread("./ex1.jpg", cv2.IMREAD_GRAYSCALE)
img_color = cv2.imread("./ex2.jpg", cv2.COLOR_BGR2RGB)
```

2. Відобразити чорно-біле та кольорове зображення.

```python
print(img_bw)
print(img_color)
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(img_bw, cmap='gray')
plt.show();
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(img_color)
plt.show();
```

3. Перетворити зображення з колірної моделі BGR в колірні моделі RGB і HSV та відобразити.

```python
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
plt.show();
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2HSV))
plt.show();
```

4. Окремо відобразити колірні канали для зображень із колірними моделями RGB та HSV.

```python
img_rgb = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 10))

plt.subplot(131);
R = np.zeros(img_rgb.shape, dtype='uint8')
R[:, :, 0] = img_rgb[:, :, 0]
plt.imshow(R)

plt.subplot(132);
G = np.zeros(img_rgb.shape, dtype='uint8')
G[:, :, 1] = img_rgb[:, :, 1]
plt.imshow(G)

plt.subplot(133);
B = np.zeros(img_rgb.shape, dtype='uint8')
B[:, :, 2] = img_rgb[:, :, 2]
plt.imshow(B)

plt.show()
```

```python
img_hsv = cv2.cvtColor(img_color, cv2.COLOR_BGR2HSV)
plt.figure(figsize=(10, 10))

plt.subplot(131);
R = np.zeros(img_hsv.shape, dtype='uint8')
R[:, :, 0] = img_hsv[:, :, 0]
plt.imshow(R)

plt.subplot(132);
G = np.zeros(img_hsv.shape, dtype='uint8')
G[:, :, 1] = img_hsv[:, :, 1]
plt.imshow(G)

plt.subplot(133);
B = np.zeros(img_hsv.shape, dtype='uint8')
B[:, :, 2] = img_hsv[:, :, 2]
plt.imshow(B)

plt.show()
```

5. Скопіювати область інтересу (Region of Interest) чорно-білого зображення, використовуючи наступні діапазони 0:img.shape[0]/2, 0:img.shape[1]/2.

```python
img_crop_bw = img_bw[0:int(img_bw.shape[0]/2), 0:int(img_bw.shape[1]/2)]
print(img_crop_bw)
plt.imshow(img_crop_bw, cmap='gray')
plt.show()
```

6. Зберегти область інтересу (Region of Interest).

```python
cv2.imwrite("./save_ex1.png", img_crop_bw)
```
